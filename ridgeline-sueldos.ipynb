```python
import pandas as pd
import matplotlib.pyplot as plt
from joypy import joyplot
from pandas.api.types import CategoricalDtype
```

```python
df = pd.read_csv('salarios_ID2.csv', sep=";", decimal="," )
print(df.head())
print(df.count())
```

```python
cat_edad  = CategoricalDtype(['MENOS 19 AOS','DE 20 A 29', 'DE 30 A 39', 'DE 40 A 49', 'DE 50 A 59', 'MS DE 59'])
df['Edad'] = df['Edad'].astype(cat_edad)
df.dtypes


plt.figure()

joyplot(
    data=df[['Salario', 'Edad']], 
    by='Edad',
    figsize=(12, 8)
)
plt.title('Salario por grupos de edad', fontsize=20)
plt.show()
```

```python
cat_estudios  = CategoricalDtype(['Menos que primaria', 'Primaria', 'Etapa 1 ESO',  'Etapa 2 ESO',  'FP', 'Licenciados', 'Diplomados'])
df['NivelEstudios'] = df['NivelEstudios'].astype(cat_estudios)

plt.figure()

joyplot(
    data=df[['Salario', 'NivelEstudios']], 
    by='NivelEstudios',
    figsize=(12, 8)
)
plt.title('Salario por grupos de Nivel de estudios', fontsize=20)
plt.show()
```

```python
cat_sexo  = CategoricalDtype(['Hombre', 'Mujer'])
df['Sexo'] = df['Sexo'].astype(cat_sexo)

plt.figure()

joyplot(
    data=df[['Salario', 'Sexo']], 
    by='Sexo',
    figsize=(12, 8)
)
plt.title('Salario por  Sexo', fontsize=20)
plt.show()
```
