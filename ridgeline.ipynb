```python
import pandas as pd
import matplotlib.pyplot as plt
from joypy import joyplot
```

```python
df = pd.read_csv('weatherAUS.csv', usecols=['Date', 'Location', 'MinTemp', 'MaxTemp'])
df.head()
```

```python
sydney = df.query("Location == 'Sydney'")
sydney = sydney.drop('Location', axis=1)
sydney['Date'] = sydney['Date'].astype('datetime64')
sydney['Month'] = sydney['Date'].dt.month_name()

sydney.head()
```

```python
from pandas.api.types import CategoricalDtype

cat_month = CategoricalDtype(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])

sydney['Month'] = sydney['Month'].astype(cat_month)

sydney.dtypes
```

Ridgeline plot for a single variable


```python
plt.figure()

joyplot(
    data=sydney[['MaxTemp', 'Month']], 
    by='Month',
    figsize=(12, 8)
)
plt.title('Ridgeline Plot of Max Temperatures in Sydney', fontsize=20)
plt.show()
```

Ridgeline plot for multiple variables


```python
plt.figure()

ax, fig = joyplot(
    data=sydney[['MinTemp', 'MaxTemp', 'Month']], 
    by='Month',
    column=['MinTemp', 'MaxTemp'],
    color=['#686de0', '#eb4d4b'],
    legend=True,
    alpha=0.85,
    figsize=(12, 8)
)
plt.title('Ridgeline Plot of Min and Max Temperatures in Sydney', fontsize=20)
plt.show()
```
